```python
directory = 'cvat-jsons'
json_dataset = os.listdir(directory)
# print(json_dataset)

result = {}

# def updateImages(newImageArray ,resultImages ):
#     for item in newImageArray:
#         lastIndex = len(resultImages)-1
#         newItem = {**item , 'id' : resultImages[lastIndex]['id'] + 1}
#         resultImages.append(newItem)
#     return resultImages

def updateAnnotations(newAnnotationsArray , resultAnnotations):
    for item in newAnnotationsArray:
        lastIndex = len(resultAnnotations)-1
        newItem = {**item , 'id' : resultAnnotations[lastIndex]['id'] + 1}
        resultAnnotations.append(newItem)
    return resultAnnotations

def updateImageData(result , currentFile):
    newImageArray = currentFile['images']
    resultImages = result['images']
    newAnnotationsArray = currentFile['annotations']
    resultAnnotations = result['annotations']
    
    for item in newImageArray:
        print(item)
        
        lastIndex = len(resultImages)-1
        newIndex = resultImages[lastIndex]['id'] + 1
        newItem = {**item , 'id' : newIndex}
        
        resultImages.append(newItem)
        
        annotations = filter(lambda obj: obj['image_id'] == item['id'] , newAnnotationsArray)
        
        for obj in list(annotations):
            if(obj['image_id'] == item['id']):
                lastIndex = len(resultAnnotations)-1
                newAnnotatedItem = {**obj , 'id' : resultAnnotations[lastIndex]['id'] + 1 , 'image_id' : newIndex}
                resultAnnotations.append(newAnnotatedItem)
        
#         
        
#         print("\n")
#         print(resultAnnotations)
        
    result['images'] = resultImages
    result['annotations'] = resultAnnotations  
    return result
        

# def mergeJson(result , currentFile):
#     print(currentFile)
#     newImageArray = currentFile['images']
#     resultImages = result['images']
#     newAnnotationsArray = currentFile['annotations']
#     resultAnnotations = result['annotations']
    
#     resultImages = updateImages(newImageArray , resultImages)
#     resultAnnotations = updateAnnotations(newAnnotationsArray , resultAnnotations)
#     result['images'] = resultImages
#     result['annotations'] = resultAnnotations
#     return result

    
    
    
```

```python
for idx, file in enumerate(json_dataset):
    f = open(directory + '/' + file,)
    currentFile = json.load(f)
    if(idx == 0):
        result = currentFile
    else:
        result = updateImageData(result , currentFile)
#         result = mergeJson(result , currentFile)
    
    

with open('annotations.json', 'w+') as json_object:
    json.dump(result, json_object)
     
```
